<h2>LOAD DATA</h2>

```python
S=pd.read_csv('data/csv/exp.p.csv')
C=pd.read_csv('data/csv/exp.c.csv')
R=pd.read_csv('data/csv/exp.r.csv')
```

<h2>EXTRACT STEP</h2>

```python
step=23
#AGENTS=S[["STEP","ID","X","Y"]].query("STEP==%d" % step)
AGENTS=S[S['STEP']==23]
#COHESION=C[["STEP","PX","PY","NX","NY",]].query("STEP==%d" % step)
COHESION=C[C["STEP"]==23]
#REPULSION=R[["STEP","PX","PY","NX","NY",]].query("STEP==%d" % step)
REPULSION=R[R["STEP"]==23]
```

```python
AGENTS.to_csv('AGENTSATSTEP23.csv')
```

```python
RELATIONS = COHESION.merge(REPULSION,how='left',on=("PID","NID"),suffixes=("_COH","_REP"))
```

```python
VECTORS = RELATIONS[["PID","PX_COH","PY_COH","NID","NX_COH","NY_COH","COHX","COHY","REPX","REPY"]].replace(np.nan, 0)
```

```python
NEXTPOS = VECTORS.sort_values(by=['PID', 'NID'])
```

```python
NEXTPOS["CHANGEX"] = (NEXTPOS["COHX"]-NEXTPOS["REPX"])
NEXTPOS["CHANGEY"] = (NEXTPOS["COHY"]-NEXTPOS["REPY"])
```

```python
NEXTPOS["NEXTX"] = (NEXTPOS["PX_COH"]+NEXTPOS["CHANGEX"])
NEXTPOS["NEXTY"] = (NEXTPOS["PY_COH"]+NEXTPOS["CHANGEY"])
```
